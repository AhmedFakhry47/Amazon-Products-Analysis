```python
#Import dependancies
import matplotlib.pyplot as plt
import random as random
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Musical_instruments_reviews.csv',delimiter=',',header=0,usecols=[0,1,2,4,5,6,8],names=['ReviewerId','ProductId','ReviewerName','ReviewHelpfullness','Text','Score','ReviewSummary','uRT','ReviewTime'])
```

# Data Query

```python
#Read Headers
data.columns

#Read Each Coloumn
data[['Text','Score']][0:2] #Select Text,Score from data; in SQLite

#Read Each Row
data.iloc[0:4]

#Read Specific Location [R,C]
#data.iloc[2,2]
data.loc[(data['ProductId'] == '1384719342') & (data['Score'] >= 4.0)] 
```

# Sorting / Describing / Grouping Data

```python
#Describing the score state of a certain product
data.loc[data['ProductId']=='1384719342'].describe()

#Sorting based on score and grouping based on ProductId
ndata = data.sort_values(by=['Score'])

#Grouping data and peforming operations on the grouped data
##Sorting Products by their scores
ndata = data.groupby(by=['ProductId'],as_index=False).mean().sort_values(by=['Score'])
```

```python
ndata.head(2)
```

```python
#Retreive the product with the lowest score and product/products with the highest score possible 5.0
## What is the lowest score

print('The lowest avg score is :', ndata['Score'].min() ,' for the following product Id :', ndata.at[ndata['Score'].idxmin(),'ProductId'])

print('Number of products with negative reviews (less than 3.0) : ', ndata['Score'].loc[ndata['Score']<=3.0].count(), '\n')

## What is the maximum score
print('The highest avg score is',ndata['Score'].max())

print('Number of products that hold 5.0 score :', ndata['Score'].loc[ndata['Score'] == 5.0].count())

print('Number of products :',ndata.count().values[0])
```

# Data Visualization
## Part A : Product Scores 

```python
#Randomize Product Id and Corresponding Score
rid,rscore = random.randint(0,20),random.randint(21,40)

plt.figure(num=1,figsize=(10,10))
plt.xticks(rotation=90)
plt.bar(ndata['ProductId'].values.flat[rid:rscore],ndata['Score'].values.flat[rid:rscore])
plt.xlabel('Product Id',fontsize=20)
plt.ylabel('Scores',fontsize=20)
plt.suptitle('Product Scores\n',fontsize=50)
plt.show()
```

```python
#Combine products by their scores 
sdata = ndata.groupby(by=['Score'],as_index=False).count().sort_values(by=['Score'])
sdata.columns = ['Score','Number']
```

```python
plt.figure(num=2,figsize=(5,5))
plt.xticks(rotation=90)
plt.xlabel('Scores',fontsize=10)
plt.ylabel('Frequency',fontsize=10)
plt.bar(sdata['Score'].values.flat,sdata['Number'].values.flat)
plt.suptitle('Score Clouds',fontsize=10)
plt.show()
```
